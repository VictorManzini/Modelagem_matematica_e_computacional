# Questão 07 (02 pontos)

Crie uma função "personalizada" em Python que aceite 2 argumentos:
- Consumo médio do veículo
- Número de litros consumidos

Como saída dessa função, ela deve devolver a distância total percorrida.

Execute essa função para valores arbitrários (mas que façam algum sentido!).

```python
def distancia(consumo, litros):
    resultado = consumo * litros
    return resultado
print(distancia(4.5, 60))
```
