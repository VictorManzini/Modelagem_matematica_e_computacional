### Exemplo

```python
from scipy.datasets import electrocardiogram
ecg = electrocardiogram()

fs = 360
time = np.arange(ecg.size) / fs
plt.plot(time, ecg)
plt.xlabel("Tempo [s]")
plt.ylabel("Sinal de ECG [mV]")
plt.xlim(15, 16)
plt.ylim(-1, 3)
plt.show()
```

```python
type(ecg)
```

```python
ecg.ndim
```

### Exemplo

```python
from PIL import Image

# Carregar a imagem local como preto e branco
img = Image.open("36808022704_72650cbb3a_c.jpg").convert("L")
img_array = np.array(img)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(img_array, cmap="gray")
plt.title("Imagem Preto e Branco")
plt.show()
```

```python
type(img_array)
```

```python
img_array.ndim
```

```python
img_array.shape
```

### Zoom no exemplo

```python
x0, y0 = 440, 104   # canto superior esquerdo da região (linha, coluna)
w, h = 10, 10       # largura e altura do quadrado
```

```python
crop = img_array[y0:y0+h, x0:x0+w]   # sub-região
```

```python
plt.figure(figsize=(6, 6))

plt.imshow(crop, cmap="gray")
plt.title(f"Sub-região {w}×{h}")

plt.xticks(range(w))
plt.yticks(range(h))

plt.show()
```

```python
crop
```

### Função personalizada

```python
def my_abs3(x):
    return np.where(x >= 0, x, -x)
```

https://numpy.org/doc/stable/reference/generated/numpy.where.html

Return elements chosen from x or y depending on condition.

```python
print(my_abs3(a))
```
