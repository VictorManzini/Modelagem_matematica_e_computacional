# Gráficos nas discontinuidades

## Exemplo 1

$f(x) = \frac{x}{|x|}$

```python
x = np.linspace(-1, 1, 1000)
y = x/np.abs(x)
```

```python
plt.figure(figsize=(4, 4))

plt.plot(x, y)
plt.title('Função sinal: $y = x/|x|$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid()

plt.show()
```

## Exemplo 2

$f(x) = \frac{x+1}{x-1}$

```python
x = np.linspace(0.9, 1.1, 100)

y = np.divide(x+1, x-1)
#y = (x+1)/(x-1)

plt.plot(x, y, label='y = x+1 / x-1')

plt.title("y = x+1 / x-1")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()

plt.grid(True)
plt.show()
```

```python
entrada1 = np.linspace(0.99, 0.99999, 100) #esquerda
saida1 = np.divide(entrada1+1, entrada1-1)

entrada2 = np.linspace(1.00001, 1.01, 100) #direita
saida2 = np.divide(entrada2+1, entrada2-1)

plt.plot(entrada1, saida1)
plt.plot(entrada2, saida2, 'r')

plt.title("y =  x+1 / x-1")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(True)
plt.show()
```

# Computação simbólica utilizando Sympy

https://docs.sympy.org/latest/tutorials/intro-tutorial/intro.html

https://docs.sympy.org/latest/tutorials/intro-tutorial/calculus.html


A computação simbólica trata da computação de objetos matemáticos simbolicamente. Isso significa que os objetos matemáticos são representados exatamente, não aproximadamente.

```python
import numpy as np
print(np.sqrt(9))
print(np.sqrt(8))
print(np.pi)
```

```python
import sympy as sp
print(sp.sqrt(9))
print(sp.sqrt(8))
print(sp.sqrt(3))
print(sp.pi)
```

Em Python, as variáveis ​​não têm significado até que sejam definidas. SymPy não é diferente. Ao contrário de muitos sistemas de manipulação simbólica que você pode ter usado, no SymPy as variáveis ​​não são definidas automaticamente. Para definir variáveis, devemos usar símbolos.

```python
x = sp.symbols('x')
print(type(x))

expr = x + 1
print(expr)
```

```python
expr = expr + sp.pi
print(expr)
```

## Fatoração e expansão de polinômios

```python
x = sp.Symbol('x')         # declara uma variável simbólica
f = x**2 + 2*x + 1         # constrói a expressão (x + 1)^2

factored = sp.factor(f)    # fatora a expressão
print(factored)  # (x + 1)**2

expanded = sp.expand(factored)    # expande a expressão
print(expanded)  # x**2 + 2*x + 1
```

## Raízes de um polinômio

```python
print(sp.solve(x**2 - sp.sqrt(8)*x, x))
```

# Limites com Sympy

Por padrão: limite lateral pela direita

https://docs.sympy.org/latest/modules/series/series.html

```python
# Exemplo 1 dos gráficos
x = sp.symbols('x')
f = sp.sign(x)
print(sp.limit(f, x, 0))

print(sp.limit(f, x, 0, dir='+'))
print(sp.limit(f, x, 0, dir='-'))
```

```python
# Exemplo 2 dos gráficos
x = sp.Symbol('x')
f = (x + 1) / (x - 1)

print(sp.limit(f, x, 1))
print(sp.limit(f, x, 1, dir='+'))
print(sp.limit(f, x, 1, dir='-'))
print(sp.limit(f, x, sp.oo))
```

```python
# Outro exemplo
x = sp.Symbol('x')
print(sp.limit(1/x, x, 0, '+'))
print(sp.limit(1/x, x, 0, '-'))
print(sp.limit(1/x, x, 0))
```
