## 05 (01 ponto)

- Defina uma variável simbólica x
- Defina f(x) = abs(x) simbolicamente:

  - https://docs.sympy.org/latest/modules/functions/elementary.html#sympy.functions.elementary.complexes.Abs

- Calcule os limites dessa função nos seguintes pontos:

  - x -> 0
  - x -> -oo
  - x -> +oo
  - x -> k (escolha uma constante k positiva não-nula)
  - x -> -k (use a mesma constante k, mas com sinal invertido)

```python
import sympy as sp

x = sp.Symbol('x')
k = 3  # constante positiva não-nula

f = sp.Abs(x)

print("Limites de f(x) = |x|:")
print(f"x -> 0:   {sp.limit(f, x, 0)}")
print(f"x -> -oo: {sp.limit(f, x, -sp.oo)}")
print(f"x -> +oo: {sp.limit(f, x, sp.oo)}")
print(f"x -> k:   {sp.limit(f, x, k)}")
print(f"x -> -k:  {sp.limit(f, x, -k)}")
```

## 06 (01 pontos)

Calcule simbólicamente com o sympy o limite de:


- (x + 5) quando x tende a 1
- (x²) quando x tende a 1
- (x² + x + 5) quando x tende a 1.

Notar que o nome das funções no sympy pode ser ligeiramente diferente do numpy. sp.Abs(), por exemplo, ao invés de np.abs()

**Ao final, responda brevemente:** O que você observou nos resultados?

```python
x = sp.Symbol('x')

f1 = x + 5
f2 = x**2
f3 = x**2 + x + 5

print(sp.limit(f1, x, 1))  # 6
print(sp.limit(f2, x, 1))  # 1
print(sp.limit(f3, x, 1))  # 7
```

## 07 (01 pontos)

  Vamos resolver simbolicamente o limite de mais duas funções

  - sqrt(x) com x tendendo a 1, depois a 0 e depois a -1.
https://docs.sympy.org/latest/modules/functions/elementary.html#sympy.functions.elementary.miscellaneous.sqrt


  - sin(1/x) com x tendendo a 1 e depois a 0.


**Ao final, responda brevemente:**
- Quais casos retornaram um número?
- Em quais casos apareceram outros tipos de mensagens?
- O que elas podem indicar?

```python
x = sp.Symbol('x')

f1 = sp.sqrt(x)
f2 = sp.sin(1/x)

print("=== sqrt(x) ===")
print(f"x -> 1:  {sp.limit(f1, x, 1)}")
print(f"x -> 0:  {sp.limit(f1, x, 0)}")
print(f"x -> -1: {sp.limit(f1, x, -1)}")

print("\n=== sin(1/x) ===")
print(f"x -> 1:  {sp.limit(f2, x, 1)}")
print(f"x -> 0:  {sp.limit(f2, x, 0)}")
```

Observações: 

sqrt(x) em x->1 e x->0, sin(1/x) em x->1 retornaram um número.
Casos especiais: sqrt(-1) retornou um número imaginário. sin(1/x) em x->0 retornou AccumBounds(-1,1).
Indicam que nem todo limite existe ou pertence aos reais.
